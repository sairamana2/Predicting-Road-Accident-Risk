```python
# import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score
import warnings

warnings.filterwarnings('ignore')

TRAIN_PATH = '/content/drive/MyDrive/Road_accident_prediction/train.csv'
TEST_PATH = '/content/drive/MyDrive/Road_accident_prediction/test.csv'
SUB_PATH = '/content/drive/MyDrive/Road_accident_prediction/sample_submission.csv'

try:
    train_df = pd.read_csv(TRAIN_PATH)
    test_df = pd.read_csv(TEST_PATH)
    sample_sub = pd.read_csv(SUB_PATH)
except FileNotFoundError:
    print("---! ERROR !---")
    print("Files not found. Please double-check your Google Drive paths.")
    print(f"Tried to load Train: {TRAIN_PATH}")
    print(f"Tried to load Test: {TEST_PATH}")

print("Data loaded successfully.")

def process_time(df):
    time_mapping = {
        'morning': 9,
        'afternoon': 14,
        'evening': 19,
        'night': 2
    }
    df['Hour'] = df['time_of_day'].map(time_mapping).fillna(0).astype(int)
    df = df.drop('time_of_day', axis=1)
    return df

train_df = process_time(train_df)
test_df = process_time(test_df)
print("Time feature processed.")

TARGET = 'accident_risk'
X = train_df.drop([TARGET, 'id'], axis=1)
y = train_df[TARGET]
X_test = test_df.drop('id', axis=1)
test_ids = test_df['id']

numerical_features = X.select_dtypes(include=np.number).columns
categorical_features = X.select_dtypes(include=['object', 'category']).columns

print(f"Found {len(numerical_features)} numerical features: {list(numerical_features)}")
print(f"Found {len(categorical_features)} categorical features: {list(categorical_features)}")

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='passthrough'
)

model = lgb.LGBMRegressor(
    random_state=42,
    n_estimators=200,
    learning_rate=0.1,
    num_leaves=31,
    n_jobs=-1
)

model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', model)
])

print("Training final model on all data...")
model_pipeline.fit(X, y)
print("Model training complete.")

print("Making predictions on the test set...")
predictions = model_pipeline.predict(X_test)
predictions = np.clip(predictions, 0, 1)

submission = pd.DataFrame({
    'id': test_ids,
    'accident_risk': predictions
})

submission.to_csv('submission.csv', index=False)

print("\n--- SUCCESS ---")
print("Submission file 'submission.csv' created successfully!")
print("\nSubmission file head:")
print(submission.head())
```
